```python
import spacy
nlp = spacy.load("en_core_web_sm")
# nlp = spacy.load("en_core_web_lg")
```

```python
def preprocess_text(text):
    doc = nlp(text)
    filtered_tokens = [token.lemma_ for token in doc
                       if not token.is_stop and    # remove stopwords
                       not token.is_punct and      # remove punctuation
                       not token.is_space and      # remove spaces
                       not token.like_num]         # remove numbers
    cleaned_text = " ".join(filtered_tokens)
    return cleaned_text

# Apply preprocessing before vectorization
df["cleaned_text"] = df["text"].apply(preprocess_text)

# # Now generate vectors from cleaned text
# df["vector"] = df["cleaned_text"].apply(lambda x: nlp(x).vector)
```

```python
df.head()
```

```python
df["Vector"] = df["cleaned_text"].apply( lambda x : nlp(x).vector)
df.head()
```

```python
X=df.Vector.values
y=df.Label.values
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
for train_idx, test_idx in split.split(X,y):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
```
